### Data Cleaning

```python
# Changed DD-MM-YYYY to YYYY-MM-DD in energy data
# changed some time from HH:MM to HH:MM:SS in energy data
```

```python
from datetime import datetime

def convert_to_datetime(value):
  return datetime.strptime(value, "%m/%d/%Y %H:%M:%S")


df_energy['Datetime'] = df_energy['Datetime'].apply(convert_to_datetime)
```

```python
type(df_energy['Datetime'][0])
```

```python
df_indoor['Datetime'] = pd.to_datetime(df_indoor['Date'] + ' ' + df_indoor['Time'])
```

```python
df_outdoor['Datetime'] = pd.to_datetime(df_outdoor['Date'] + ' ' + df_outdoor['Time'])
```
